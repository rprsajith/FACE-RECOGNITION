```python
import face_recognition
import cv2
from google.colab.patches import cv2_imshow
```

```python
image1 = face_recognition.load_image_file('/content/virat kohli 1.jpg')
sajith_encoding = face_recognition.face_encodings(image1)[0]
known_encoding = [sajith_encoding]
known_names = ['virat kohli']
```

```python
known_encoding
```

```python
image2 = face_recognition.load_image_file('/content/virat kohli 2.jpg')
image2_encoding = face_recognition.face_encodings(image2)[0]
image2_encoding
```

```python
results = face_recognition.compare_faces(known_encoding,image2_encoding)
results
```

```python
for i in range(len(results)):
  if results[i]:
    name = known_names[i]
    (top,left,bottom,right)=face_recognition.face_locations(image2)[0]
    cv2.rectangle(image2,(left,top),(right,bottom),(0,0,255),2)
    cv2.putText(image2,name,(left+2,bottom+20),cv2.FONT_HERSHEY_PLAIN,1,(255,255,255),1)
    cv2_imshow(image2)
```
